# Categorical

```python
df.fillna(df.mean(numeric_only=True),inplace=True)
df.head()
```

```python
for col in df1.select_dtypes(include='object'):
  df1[col]=df1[col].fillna(df1[col].mode()[0])
df1.head()
```

#MERGE_DATASETS

```python
Customer = pd.merge(df,df1,
                    left_on='Customer_ID',right_on='CustomerID',how="left")
Customer.head()
```

#AGE_GROUP


```python
Customer['Age_Group']=pd.cut(Customer['Age_x'],
          bins=[0,30,45,60,75],
          labels=["Younger","ADULTS","MID-ADULT","SENIOR"])
Customer.head()
```

#FREQUENCY_VISTORS_MAPPING

```python
Feq_map ={
    "Weekly":4,"Bi-Weekly":4,
    "Monthly":12,"Quarterly":48,
    "Yearly":5216,"Annually": 2,
    "Fortnightly":3,"Every 3 Months":3

}
Customer['FrequencyOfPurchases']=Customer['FrequencyOfPurchases'].map(Feq_map)
Customer.head()
```

#TOTAL_SPENDING



```python
Customer['Purchase_Amount_USD']=Customer.groupby("Customer_ID")["Purchase_Amount_USD"].transform(lambda x:x.sum())
Customer.head()
```

#CUSTOMER SEGMENTATION(K_MEANS)


```python
# Select the features for clustering
features_Customer = Customer[["Age_Group", "FrequencyOfPurchases", "Purchase_Amount_USD"]]

# One-hot encode the 'Age_Group' categorical column
features_Customer = pd.get_dummies(features_Customer, columns=['Age_Group'], drop_first=True)

# Scale the numerical features
scaler = StandardScaler()
Features_scaled = scaler.fit_transform(features_Customer)

# Initialize and fit the KMeans model
KMEANS = KMeans(n_clusters=4, random_state=42, n_init=10) # Added n_init to suppress warning
KMEANS.fit(Features_scaled)
```

```python
Customer['Cluster'] = KMEANS.labels_

fig = px.scatter_3d(Customer, x='FrequencyOfPurchases', y='Purchase_Amount_USD', z='Age_Group',
                    color='Cluster', title='Customer Segments by Purchase Behavior and Age Group')
fig.show()
```

#SEGMENT EVALUATION_Frequency_Purchase

```python
# Calculate silhouette score
silhouette_AVG = silhouette_score(Features_scaled, KMEANS.labels_)
print("Silhouette Score:", silhouette_AVG)
```

```python
x=features_Customer
y=Customer["FrequencyOfPurchases"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=DecisionTreeClassifier()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
print(classification_report(y_test,y_pred))
```
